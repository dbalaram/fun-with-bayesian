### Variational Inference

#### Toy Problem: Predicting Tennis Match Win Probability using Variational Inference


P(θ∣data)∝P(data∣θ)P(θ)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
def variational_inference_tennis(wins, losses, prior_alpha, prior_beta, num_iterations=10):
    alpha = prior_alpha
    beta_val = prior_beta
    alphas = []
    betas = []
    probabilities = []

    for i in range(num_iterations):
        alpha += wins
        beta_val += losses
        
        alphas.append(alpha)
        betas.append(beta_val)
        prob = alpha / (alpha + beta_val)
        probabilities.append(prob)
        print(f"Iteration {i+1}: Win Probability = {prob:.4f}")

    iterations = np.arange(1, num_iterations + 1)
    plt.figure(figsize=(10, 5))
    plt.plot(iterations, probabilities, label='Win Probability')
    plt.title('Tennis Match Win Probability Update')
    plt.xlabel('Iteration')
    plt.ylabel('Win Probability')
    plt.legend()
    plt.show()
```

```python
total_matches = 15
wins = np.random.binomial(1, 0.75, total_matches).sum()  # Simulate with a true win probability of 70%
print (wins)
losses = total_matches - wins
```

```python

variational_inference_tennis(wins, losses, prior_alpha=1, prior_beta=1)
```
